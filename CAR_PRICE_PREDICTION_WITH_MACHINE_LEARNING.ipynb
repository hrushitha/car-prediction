```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/CarPrice.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
x = pd.DataFrame({'Horse Power' : df['horsepower'], 'Peak RPM': df['peakrpm'],'City MPG ': df['citympg'],'Highway mpg ': df['highwaympg']})
```

```python
y = np.array(df['price'])
```

```python
x
```

```python
y
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 ,random_state = 5000)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_r = LinearRegression()
linear_r.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
y_pred = linear_r.predict(x_test)
print('R2 of linear Regresison : ', r2_score(y_test,y_pred))
print('mse of Linear Regression : ',mean_squared_error(y_test,y_pred))
print('mae of Linear Regression : ',mean_absolute_error(y_test,y_pred))
```

```python
import pickle
filename = 'lr_model.sav'
pickle.dump(linear_r,open(filename,'wb'))
```
